```python
df = pd.read_csv("titanic.csv")

df_male = df[df["Sex"] == "male"]
df_female = df[df["Sex"] == "female"]

df_male
df_female

df_female = df_female.dropna() # в созданных срезах удаляем строки, где есть пропуски
df_male = df_female.dropna()
```

```python
plt.hist(df_male["Age"], bins=12, label="Мужчины") # создаём гистрограммы для каждого среза
plt.xlabel("Возраст")
plt.ylabel("Количество пассажиров")
plt.title("Распределение пассажиров по возрасту (Мужчины)")
plt.grid(True)
plt.show()
```

```python
plt.hist(df_female["Age"], bins=12, label="Женщины")
plt.xlabel("Возраст")
plt.ylabel("Количество пассажиров")
plt.title("Распределение пассажиров по возрасту (Женщины)")
plt.grid(True)
plt.show()
```
